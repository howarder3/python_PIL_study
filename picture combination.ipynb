```python
from PIL import Image
from matplotlib.pyplot import imshow
from IPython.display import display # to display images
from glob import glob

#opens images:
filelist = glob("./*.jpg")
print("find file: {}".format(filelist))

result_width, result_height = 560, 560
result_im = Image.new('RGB', (result_width, result_height))

x, y = 0, 0
step_size = 56

for file in filelist:
    im = Image.open(file)
    print(im.size)
#     im.thumbnail((step_size,step_size))
    im = im.resize((step_size, step_size), Image.BILINEAR )
    result_im.paste(im, (x,y))
    if x >= result_width:
        x = 0
        y = y + step_size
    elif y >= result_width:
        break
    else:
        x = x + step_size   
        
display(result_im)
result_im.save("result.jpg", quality=100)
```
